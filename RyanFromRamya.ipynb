```python
#this creates a new df that includes the regional indicator

data2 = pd.read_csv(f"combined_happiness_after_clean.csv")

combined_with_regional_indicator = pd.merge(Combined, data2, on='Country', how='left')

combined_with_regional_indicator.head()
```

```python
#This creates a df that gets the mean of each region for each category

new_df = combined_with_regional_indicator.groupby("Regional indicator").agg("mean")
new_df.head(20)
```

```python
Region_Rank_df = new_df.iloc[:,30:55]
Region_Rank_df.head(2)
```

```python
#GDP_Mean = Region_Rank_df[["GDP 2017 Ranking", "GDP 2018 Ranking", "GDP 2019 Ranking", "GDP 2020 Ranking", "GDP 2021 Ranking"]].mean()
#GDP_Mean['avg'] = Region_Rank_df[["GDP 2017 Ranking", "GDP 2018 Ranking"]].mean(axis=1)

GDP_Mean = Region_Rank_df.assign(GDP_Average=Region_Rank_df.loc[:, ["GDP 2017 Ranking", "GDP 2018 Ranking", 
                                                                    "GDP 2019 Ranking", "GDP 2020 Ranking", 
                                                                    "GDP 2021 Ranking"]].mean(axis=1))

Life_Mean = GDP_Mean.assign(Life_Average=Region_Rank_df.loc[:, ["Life Expectancy Ranking 2017", 
                                                                                "Life Expectancy Ranking 2018", 
                                                                                "Life Expectancy Ranking 2019", 
                                                                                "Life Expectancy Ranking 2020", 
                                                                                "Life Expectancy Ranking 2021"]].mean(axis=1))

Ladder_Mean = Life_Mean.assign(Ladder_Average=Region_Rank_df.loc[:, ["Ladder Score Ranking 2017", 
                                                                                "Ladder Score Ranking 2018", 
                                                                                "Ladder Score Ranking 2019", 
                                                                                "Ladder Score Ranking 2020", 
                                                                                "Ladder Score Ranking 2021"]].mean(axis=1))

Freedom_Mean = Ladder_Mean.assign(Freedom_Average=Region_Rank_df.loc[:, ["Freedom Ranking 2017", 
                                                                                "Freedom Ranking 2018", 
                                                                                "Freedom Ranking 2019", 
                                                                                "Freedom Ranking 2020", 
                                                                                "Freedom Ranking 2021"]].mean(axis=1))

Generosity_Mean = Freedom_Mean.assign(Generosity_Average=Region_Rank_df.loc[:, ["Generosity Ranking 2017", 
                                                                                "Generosity Ranking 2018", 
                                                                                "Generosity Ranking 2019", 
                                                                                "Generosity Ranking 2020", 
                                                                                "Generosity Ranking 2021"]].mean(axis=1))

Combined_Mean = Generosity_Mean

#This takes isolates the columns that display the averages

Combined_Mean = Combined_Mean.iloc[:,25:30]

Combined_Mean.head(10)
```

```python
Combined_Mean.plot.scatter(x = 'GDP_Average', y = 'Ladder_Average')

Combined_Mean.plot.scatter(x = 'Life_Average', y = 'Ladder_Average')

Combined_Mean.plot.scatter(x = 'Freedom_Average', y = 'Ladder_Average')

Combined_Mean.plot.scatter(x = 'Generosity_Average', y = 'Ladder_Average')

```
